```python
# import pandas
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Carpentries link for gapminder data
data_url = 'http://bit.ly/2cLzoxH'
#load gapminder data from url as pandas dataframe
gapminder = pd.read_csv(data_url)
print(gapminder.head(3))
```

```python
gapminder_us = gapminder[gapminder.country=="United States"]
```

```python
# create figure and axis objects with subplots()
fig,ax= plt.subplots()
ax.plot(gapminder_us.year, gapminder_us.lifeExp, marker="o")
ax.set_xlabel("year")
ax.set_ylabel("lifeExp")
ax.plot(gapminder_us.year, gapminder_us["gdpPercap"], marker="o")
plt.show()
```

```python
def plot_2_y_axes(data1,data2,x_label,y_label1,y_label2):
    plt.figure(7)
    # create figure and axis objects with subplots()
    fig,ax = plt.subplots()
    # make a plot
    ax.plot(data1,color="red",marker="o")
    # set x-axis label
    ax.set_xlabel(x_label)
    # set y-axis label
    ax.set_ylabel(y_label1)
    # twin object for two different y-axis on the sample plot
    ax2=ax.twinx()
    # make a plot with different y-axis using second axis object
    ax2.plot(data2,color="blue",marker="o")
    ax2.set_ylabel(y_label2)
    plt.show()
```

```python
# create figure and axis objects with subplots()
fig,ax = plt.subplots()
# make a plot
ax.plot(gapminder_us.year,gapminder_us.lifeExp,color="red",marker="o")
# set x-axis label
ax.set_xlabel("year")
# set y-axis label
ax.set_ylabel("lifeExp")
# twin object for two different y-axis on the sample plot
ax2=ax.twinx()
# make a plot with different y-axis using second axis object
ax2.plot(gapminder_us.year, gapminder_us["gdpPercap"],color="blue",marker="o")
ax2.set_ylabel("gdpPercap")
plt.show()

plot_2_y_axes(gapminder_us.lifeExp,gapminder_us["gdpPercap"],'year','lifeExp','gdpPerCap')
```
